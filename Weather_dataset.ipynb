# **Data Dictionary**

* Date/Time ---> Date of weather happening something different
* Temp_C ---> Temperature
* Dew Point Temp_c ---> The temperature to which air must be cooled to become saturated with water vapor, assuming constant air pressure and water content
* Rel Hum_% ---> Humidity is the concentration of water vapor present in the air
* Wind Speed_km/h --->  Wind speed in kilometer per hour
* Visibility Km ---> Visibility is the distance than can be seen in daylight

* Press_Kpa ---> Pressure and stress in the metre-kilogram-second system
* Weather ---> Weathers name decided by units
