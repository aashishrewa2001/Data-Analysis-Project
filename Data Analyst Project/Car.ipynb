# **Data Dictionary**

* **Make** --> The brand or manufacturer of the car.

* **Model** --> The specific model name of the car.

* **Type** --> The body type or style of the car, such as sedan, coupe, SUV, etc.

* **Origin** --> The country of origin for the car's manufacturer.

* **DriveTrain** --> The type of drivetrain for the car, such as front-wheel drive (FWD), rear-wheel drive (RWD), all-wheel drive (AWD), or four-wheel drive (4WD).

* **MSRP** --> The manufacturer's suggested retail price for the car.

* **Invoice** --> The invoice price, or the price that the dealer pays the manufacturer for the car.

* **EngineSize** --> The size of the car's engine, typically measured in liters (L) or cubic centimeters (cc).

* **Cylinders** --> The number of cylinders in the car's engine.

* **Horsepower** --> The power of the car's engine, typically measured in horsepower (hp).

* **MPG_City** --> The estimated miles per gallon (MPG) the car can achieve in city driving conditions.

* **MPG_Highway** --> The estimated miles per gallon (MPG) the car can achieve in highway driving conditions.

* **Weight** --> The weight of the car, typically measured in pounds (lbs).

* **Wheelbase** --> The distance between the centers of the front and rear wheels of the car, typically measured in inches (in).

* **Length**--> The length of the car, typically measured in inches (in).
