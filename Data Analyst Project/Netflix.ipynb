# **NetFlix Data Analysis**

This Netflix Dataset has information about the TV Shows and Movies available on Netflix 2021

# **Data Dictionary**

* **Show_Id**  ---> A unique identifier for each TV show or movie in the dataset.
* **Category** ---> The category or genre of the TV show or movie, such as "TV Shows," "Movies,"
* **Title**  --->  The title or name of the TV show or movie.
* **Director** ---> The name of the director(s) of the TV show or movie
* **Cast** --->  The name of the actors/actresses in the TV show or movie.
* **Country** --> The country or countries where the TV show or movie was produced. 
* **Release_date** ---> The date when the TV show or movie was released on Netflix.
* **Rating**  ---> The maturity rating or age rating of the TV show or movie, such as "TV-MA" (Mature Audiences), "PG-13," "R," etc.
          
      "R"  --->The Movie contains adult material such as adult activity,   
                harsh language  intense graphic violence, drug abuse and nudity.

      "TV-MA"  -->Mature Audience Only. Intended for adults and may be 
                  unsuitable for children under 17.

      "TV 14"  --->The content of the program may be inappropriate for  
                   children under the age of 14, and parental guidance is strongly suggested.

      "PG-13"  ----> PARENTS STRONGLY CAUTIONED

      "NR"  ----> Not Rated

      "TV-Y"  ---->Indicates that programming is designed for chil- dren ages 7 
                   and older
                  
      "TV-G" --->  Programs suitable for all ages

      "PG" ---> Parental guidance suggested 
  
* **Duration** ---> The total duration of the TV show or movie, typically measured in minutes.

* **Type**  ---> Whether the entry is a TV show or a movie

* **Description** --->A brief synopsis or summary of the TV show or movie.
