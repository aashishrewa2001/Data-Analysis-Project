# **Data Dictionary**
* **acousticness** ---> A confidence measure from 0.0 to 1.0 of whether the track is acoustic.
* **danceability** --->Danceability is measured using a mixture of song features 
* **duration_ms** --->The duration of the track in milliseconds.
* **energy** ---> Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. 
* **instrumentalness** ---> How likely the music contains no spoken word vocals.
* **key** ---> Key is an absolute integer value denoting the scale of the song 
* **liveness** --->Detects the presence of an audience in the recording
* **loudness** ---> A loudness reference for normalising the loudness of a track
* **mode** ---> Indicates whether a track is in a major or minor key (0 = minor, 1 = major).
* **speechiness**---> A measure from 0.0 to 1.0 that detects the presence of spoken words in a track.
* **tempo** ---> This directly relates to the song's beats per minute (BPM) in the song, with higher numbers typically meaning a more energetic tune.
* **time_signature** ---> Indicate how many beats are in each measure of a piece of music, as well as which note value is counted as a beat
* **valence** --->A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track.
* **target** ---> A binary variable indicating whether the track is popular or not (**Popular** =1 , **Not Popular** =0
* **song_title** ---> The title of the track.
* **artist** ---> The name of the artist who performed the track.

# 1. What are top 5 artists with the most songs in the dataset ?

```python
df["artist"].value_counts()
```

```python
top_artist = df.groupby("artist").size().sort_values(ascending = False)[0:5]
top_artist
```

```python



plot=plt.bar(top_artist.index,top_artist.values)
plt.title("Top 5 artists with the most songs")
plt.xlabel("Number Of Song")
plt.grid()
for p in plot.patches:
  height=p.get_height()
  plt.annotate(f"{height}",xy=(p.get_x()+p.get_width()/2,height),xytext=(0,2),
             textcoords="offset points",ha="center",va="bottom")
plt.xticks(rotation=30)
plt.ylabel("Top 5 Artists")
plt.show()
```

```python
bar_value = sns.barplot(x=top_artist.index,y=top_artist.values)
for p in bar_value.patches:
  height=p.get_height()
  plt.annotate(f"{height}",xy=(p.get_x()+p.get_width()/2,height),xytext=(0,2),
               textcoords="offset points",ha="center",va="bottom")
plt.xticks(rotation=20)
```

# 3. What is the distribution of energy levels in the dataset? Are there any outliers?


```python
q1 = df["energy"].quantile(0.25)
q3 = df["energy"].quantile(0.75)
iqr = q3-q1
iqr
```

```python
outliers = df[(df["energy"] < q1 - 1.5*iqr) | (df["energy"] > q3 - 1.5*iqr)]
len(outliers)
```

```python
print(f"No of outliers = {len(outliers)}")
```
