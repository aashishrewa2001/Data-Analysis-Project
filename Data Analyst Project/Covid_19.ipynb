# Data Dictionary

* **Date** ---> The date when the data was recorded
* **State** ---> The name of the state or province where the data was recorded
* **Region** -->  This column represent the name of the country or geographic region where the data was recorded
* **Corfirmed** --> The number of confirmed cases of COVID-19 in the state or region as of the date recorded
* **Deaths** ---> The number of deaths attributed to COVID-19 in the state or region as of the date recorded.
* **Recovered** --->The number of people who have recovered from COVID-19 in the state or region as of the date recorded
